```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Replace 'salary.csv' with your actual file path if needed
data = pd.read_csv('./PR_EXAM_DATASET/Salary_Data.csv')
print(data.head())
```

```python
X = data[['YearsExperience']]
y = data['Salary']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
experience_to_predict = np.array([[15]])
predicted_salary = model.predict(experience_to_predict)
print(f"Predicted salary for 15 years of experience: ${predicted_salary[0]:,.2f}")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(15, predicted_salary, color='green', s=100, label='Prediction (15 yrs)')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
