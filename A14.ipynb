```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from seaborn
iris = sns.load_dataset('iris')

# Display first few rows
print(iris.head())
```

```python
print("\nFeature Types:")
for column in iris.columns:
    dtype = iris[column].dtype
    feature_type = "Numeric" if pd.api.types.is_numeric_dtype(iris[column]) else "Nominal"
    print(f"- {column}: {dtype} → {feature_type}")
```

```python
iris.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histograms of Iris Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Create box plots grouped by species
for column in iris.columns[:-1]:  # Skip 'species'
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f"Box Plot of {column} by Species")
    plt.show()
```
