```python
# Step 1: Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
# Step 2: Load Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Display the first few rows to understand the structure
print(titanic_data.head())

```

```python
# Step 3: Plot histogram of 'fare'
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='fare', bins=30, kde=True, color='skyblue')
```

```python

# Step 4: Set titles and labels
plt.title('Distribution of Titanic Ticket Prices (Fare)', fontsize=16)
plt.xlabel('Fare Price', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
```

```python

# Step 5: Show plot
plt.grid(True)
plt.show()
```
